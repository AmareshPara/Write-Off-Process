## Non-EDC Loans Write Off

```python
nonEDC_Loan_DF = []
before_date = pd.Timestamp(2020,3,1)
#financingTransferDate = parser.parse(row['expected financing transfer date'])
```

```python
for index,row in loansDF_US.iterrows(): # non-EDC loans that have 'Is Not Shopify Payments Restricted' don't qualify
    if (row['expected financing transfer date'] < before_date) or (row['shop province/state'] in ['Indiana','South Carolina','Washington','Puerto Rico']) or (row['is shopify payments restricted'] != 'Is Shopify Payments Restricted'):
        nonEDC_Loan_DF.append(row) 
```

## EDC Loans

```python
EDC_Loan_DF = []

after_date = pd.Timestamp(2020,2,29)

  
```

```python
aftDate = pd.Timestamp(2020,12,31)
befDate = pd.Timestamp(2020,3,1)

SP_filter = mcaDF_US['is shopify payments restricted'] == 'Is Shopify Payments Restricted'
US_MCA_DF = mcaDF_US[SP_filter]

pRico_filter = US_MCA_DF['shop province/state'] != 'Puerto Rico'
allUS_MCA_DF = US_MCA_DF [pRico_filter]
```

```python
p3_floorDate = pd.Timestamp(2019,1,1)
p3_ceilDate = pd.Timestamp(2019,12,31)

p4_floorDate = pd.Timestamp(2020,1,1)
p4_ceilDate = pd.Timestamp(2020,12,31)

p5_floorDate = pd.Timestamp(2021,1,1)
p5_ceilDate = pd.Timestamp(2022,3,31)


for index,row in EDC_MCA_Data.iterrows():
    if (row['expected financing transfer date'] >= p3_floorDate and row['expected financing transfer date'] <= p3_ceilDate ):
        EDC_MCA_Data.at[index,'EDC Policy'] = 'Policy 3'
    
    if (row['expected financing transfer date'] >= p4_floorDate and row['expected financing transfer date'] <= p4_ceilDate ):
        EDC_MCA_Data.at[index,'EDC Policy'] = 'Policy 4'
        
    if (row['expected financing transfer date'] <= p5_ceilDate and row['expected financing transfer date'] >= p5_floorDate ):
        EDC_MCA_Data.at[index,'EDC Policy'] = 'Policy 5'
```
